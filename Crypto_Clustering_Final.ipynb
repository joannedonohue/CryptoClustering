```python
# Read data into Data Frame
market_data_df = pd.read_csv("crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python

# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_scaled = StandardScaler().fit_transform(market_data_df)

# View Results
#crypto_scaled
```

```python
# Create a DataFrame with the scaled data
crypto_transformed_df = pd.DataFrame(crypto_scaled, columns=market_data_df.columns)
crypto_transformed_df.head()

# Copy the crypto names from the original data
x = market_data_df.index.values
x = pd.DataFrame(x, columns=['coin_id'])

# Set the coinid column as index
crypto_transformed_df = crypto_transformed_df.reset_index(drop=True)
crypto_transformed_df = pd.concat([crypto_transformed_df, x], axis = 1)
crypto_transformed_df = crypto_transformed_df.set_index('coin_id')

# Display sample data
crypto_transformed_df.head()
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_transformed_df)
    inertia.append(km.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow.head(10)
```

```python
# Because our data is unlabled, we will use the Elbow Method to find the optimal number of clusters (value of k)

# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Inertia is used as the objective function, it's the sum of squared distances of samples to their closest cluster center

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_transformed_df)
    inertia.append(km.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow.head(10)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.plot.line(x = "k",
                   y = "inertia",
                   title = "Elbow Curve",
                   xticks = k)
```

#### Answer the following question:
**Question:** What is the best value for `k`?

**Answer:** Based on the Elbow Curve, the optimial number of clusters is around k=4. After this point, the reduction in inertia slows down significantly, suggesting that increasing the number of clusters beyond 4 does not improve the model significantly.

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters = 4, n_init = 'auto', random_state = 0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(crypto_transformed_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model.predict(crypto_transformed_df)

# View the resulting array of cluster values.
k_4
```

### Optimize Clusters with Principal Component Analysis.

```python
# Now we'll try the PCA model approach, to reduce the feature to three principal componenets

# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
crypto_pca = pca.fit_transform(crypto_transformed_df)

# View the first five rows of the DataFrame.
crypto_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information  can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question:

**Question:** What is the total explained variance of the three principal components?

**Answer:** To find the total explained variance of the three principal components, we sum up the explained variance ratios:

0.3719856 + 0.34700813 + 0.17603793 = 0.89503166

Approx. 89.5%, meaning they capture almost 90% of the total variance in the dataset.

```python
# Create a new DataFrame with the PCA data.
# Note: The code for this step is provided for you

# Creating a DataFrame with the PCA data
pca_df = pd.DataFrame(
    data = crypto_pca, columns = ["PCA1", "PCA2", "PCA3"]
)
pca_df.head(10)

# Copy the crypto names from the original data
x = market_data_df.index.values
x = pd.DataFrame(x, columns=['coin_id'])

# Set the coinid column as index
pca_df = pca_df.reset_index(drop=True)
pca_df = pd.concat([pca_df, x], axis = 1)
pca_df = pca_df.set_index('coin_id')

# Display sample data
pca_df.head(10)
```
